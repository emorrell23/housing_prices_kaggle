```python
y = df['SalePrice'].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
Xs = ss.fit_transform(X)
```

```python
Xs
```

# Train Test Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(Xs, y)
```

# Gradient Boost Model

```python
# run a gradient boost model with gridsearch
gb = GradientBoostingRegressor()
gb_params = {
    'n_estimators': [75, 100, 125, 150],
    'max_depth': range(2,6),
    'max_features': ['auto', 'log2', None],
#     'min_samples_split': [1,2,3,4]
}
gb_model = GridSearchCV(gb, param_grid=gb_params, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
gb_model.fit(X_train, y_train)
print('best score', gb_model.best_score_)
print('best params', gb_model.best_params_)
print('test score', gb_model.score(X_test, y_test))
gb_best = gb_model.best_estimator_
```

# Random Forest Model

```python
rfr = RandomForestRegressor()
rfr_params = {
    'n_estimators': [500,1000],
    'max_depth': [None,2,4,6],
    'max_features': [None, 'sqrt', 'log2'],
    'min_samples_split': [3,5]
}
rfr_model = GridSearchCV(rfr, param_grid=rfr_params, scoring='neg_mean_squared_error', cv=4, n_jobs=-1, verbose=1)
rfr_model.fit(X_train, y_train)
print('best score', rfr_model.best_score_)
print('best params', rfr_model.best_params_)
print('test score', rfr_model.score(X_test, y_test))
rfr_best_model = rfr_model.best_estimator_
# (n_estimators=10, criterion='mse', max_depth=None, min_samples_split=2, min_samples_leaf=1, 
# min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, min_impurity_decrease=0.0, 
# min_impurity_split=None,bootstrap=True, oob_score=False, n_jobs=1, random_state=None, verbose=0, warm_start=False)
```
